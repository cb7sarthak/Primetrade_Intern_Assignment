```python
import pandas as pd
import os

# 1. Finding your files automatically
all_files = os.listdir()
sentiment_file = [f for f in all_files if 'fear' in f.lower()][0]
trades_file = [f for f in all_files if 'historical' in f.lower()][0]

print(f"Opening files: {sentiment_file} and {trades_file}")

# 2. THE SUPER LOAD: Try reading as Excel first, then as CSV
try:
    # Try reading as an Excel file first
    sentiment = pd.read_excel(sentiment_file)
    print("Success: Loaded Sentiment as Excel")
except:
    # If Excel fails, try reading as CSV
    sentiment = pd.read_csv(sentiment_file, on_bad_lines='skip', encoding='latin1')
    print("Success: Loaded Sentiment as CSV")

try:
    # Load the trading data
    trades = pd.read_csv(trades_file, on_bad_lines='skip', encoding='latin1')
    print("Success: Loaded Trades as CSV")
except:
    trades = pd.read_excel(trades_file)
    print("Success: Loaded Trades as Excel")

# 3. Show the results
print("\n--- Fear & Greed Data ---")
display(sentiment.head())

print("\n--- Trading Data ---")
display(trades.head())
```

```python
trades.head()
```

```python
print("Sentiment Shape:", sentiment.shape)
print("Trades Shape:", trades.shape)

print("\nMissing Values:")
print(sentiment.isnull().sum())
print(trades.isnull().sum())

print("\nDuplicates:")
print(sentiment.duplicated().sum())
print(trades.duplicated().sum())
```

```python
# 1. Fix Sentiment Dates (Simple conversion)
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

# 2. Fix Trades Dates with the "Day First" rule
# This tells Python that 18-03 means March 18th, not a fake 18th month!
trades['Timestamp IST'] = pd.to_datetime(
    trades['Timestamp IST'], 
    dayfirst=True, 
    format='mixed'
)

# 3. Create the clean date column for merging
trades['date'] = trades['Timestamp IST'].dt.date

# 4. Success Check
print("Success! Dates are now properly aligned.")
display(trades[['Timestamp IST', 'date']].head())
```

```python
sentiment['classification'] = sentiment['classification'].replace({
    'Extreme Fear': 'Fear',
    'Extreme Greed': 'Greed'
})
```

```python
merged = trades.merge(
    sentiment[['date','classification']],
    on='date',
    how='left'
)

merged.head()
```

```python
# 1. Define the exchange rate
usd_to_inr = 90.79

# 2. Convert Profit and Position Size to INR
merged['Closed PnL INR'] = merged['Closed PnL'] * usd_to_inr
merged['Size INR'] = merged['Size USD'] * usd_to_inr

# 3. Quick check to see the conversion
print("--- Conversion Successful (INR) ---")
display(merged[['Coin', 'Closed PnL', 'Closed PnL INR', 'Size USD', 'Size INR']].head())
```

```python
# 1. First, create the 'win' column properly
# We use the exact name 'Closed PnL' from your file
merged['win'] = merged['Closed PnL'] > 0

# 2. Average Profit per Market Mood
pnl_by_sentiment = merged.groupby('classification')['Closed PnL'].mean()
print("--- Average Profit by Sentiment ---")
print(pnl_by_sentiment)

# 3. Win Rate per Market Mood
# Now that 'win' is created above, this line will work!
win_rate = (merged.groupby('classification')['win'].mean() * 100).round(2)
print("\n--- Win Rate (%) by Sentiment ---")
print(win_rate)

# 4. Total Number of Trades per Market Mood
trade_count = merged['classification'].value_counts()
print("\n--- Total Trade Count ---")
print(trade_count)
```

```python
# Grouping by sentiment to see Average Profit, Win Rate, and Total Trades
final_report = merged.groupby('classification').agg({
    'Closed PnL INR': 'mean',
    'win': 'mean',
    'Coin': 'count'
}).rename(columns={'win': 'Win Rate (%)', 'Coin': 'Total Trades', 'Closed PnL INR': 'Avg Profit (₹)'})

# Formatting for a clean look
final_report['Win Rate (%)'] = (final_report['Win Rate (%)'] * 100).round(2)
final_report['Avg Profit (₹)'] = final_report['Avg Profit (₹)'].round(2)

print("--- FINAL TRADING PERFORMANCE REPORT (INR) ---")
display(final_report)
```

### Insight 1: PnL Distribution vs. Market Sentiment
**Analysis:** The boxplot below visualizes the spred of profits and losses. m
- **Observation:** We see wider "whiskers" during Greed days, indicating higher volatility.
- **Reasoning:** Traders tend to take more aggressive risks when the market is euphoric.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Fix: Added hue='classification' and legend=False to satisfy the update
sns.boxplot(
    x='classification', 
    y='Closed PnL', 
    data=merged, 
    hue='classification', 
    palette="viridis", 
    legend=False
)

plt.title("Profit & Loss Distribution: Fear vs Greed", fontsize=15)
plt.xlabel("Market Mood (Sentiment)", fontsize=12)
plt.ylabel("Profit / Loss ($)", fontsize=12)
plt.axhline(0, color='red', linestyle='--', linewidth=1.2, label='Break Even')

plt.show()
```

### Insight 2: Win Rate Consistency
**Analysis:** This chart compares the percentage of successful trades across different market moods.
- **Observation:** The win rate is notably higher during **Fear** days compared to **Greed**.
- **Reasoning:** During Fear, the trader likely waits for stronger "bottom" signals, whereas Greed leads to "chasing" price action (FOMO).
- **Actionable Rule:** High-confidence entries are more frequent in Fear; maintain discipline during Greed.

### Insight 3: Trade Segmentation & Risk Behavior
**Analysis:** We categorized trades into "Big" vs "Small" based on the median position size.
- **Observation:** Our analysis shows that "Big Trades" (High Risk) often have a lower average win rate than "Small Trades."
- **Reasoning:** This suggests that as the "stakes" increase, trader psychology may be negatively impacted, leading to poorer decision-making or premature exits.
- **Strategic Rule:** To scale the account, the trader should work on "Risk Normalization"—treating big trades with the same mechanical discipline as small ones.

```python
# 1. Calculate Average Position Size by Sentiment in INR
# Using the 'Size INR' column we created earlier
size_by_sentiment_inr = merged.groupby('classification')['Size INR'].mean()

# 2. Create the bar chart
plt.figure(figsize=(8, 6))
size_by_sentiment_inr.plot(kind='bar', color='purple', edgecolor='black')

# 3. Add labels and styling
plt.title("Average Position Size (INR) by Market Sentiment", fontsize=14)
plt.xlabel("Market Mood", fontsize=12)
plt.ylabel("Average Size (₹)", fontsize=12)
plt.xticks(rotation=45)

# 4. Add data labels (formatted for Rupees)
for i, v in enumerate(size_by_sentiment_inr):
    # This adds the ₹ symbol and comma separators for thousands
    plt.text(i, v + (v * 0.05), f"₹{v:,.2f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Grouping by sentiment to see Average Profit, Win Rate, and Total Trades
final_report = merged.groupby('classification').agg({
    'Closed PnL INR': 'mean',
    'win': 'mean',
    'Coin': 'count'
}).rename(columns={'win': 'Win Rate (%)', 'Coin': 'Total Trades', 'Closed PnL INR': 'Avg Profit (₹)'})

# Formatting for a clean look
final_report['Win Rate (%)'] = (final_report['Win Rate (%)'] * 100).round(2)
final_report['Avg Profit (₹)'] = final_report['Avg Profit (₹)'].round(2)

print("--- FINAL TRADING PERFORMANCE REPORT (INR) ---")
display(final_report)
```

```python
import numpy as np

# 1. We use 'Size INR' since that's your actual risk measure
median_size = merged['Size INR'].median()

# 2. Create the segment based on the Median
merged['trade_segment'] = np.where(
    merged['Size INR'] > median_size, 
    'Big Trade (High Risk)', 
    'Small Trade (Low Risk)'
)

# 3. Check how many trades fall into each
print(f"Median Trade Size: ₹{median_size:,.2f}")
print(merged['trade_segment'].value_counts())
```

```python
# 1. Group by the new segment
segment_summary = merged.groupby('trade_segment').agg({
    'win': 'mean',
    'Closed PnL INR': ['mean', 'sum']
})

# 2. Flatten column names for easier printing
segment_summary.columns = ['Win Rate (%)', 'Avg Profit (₹)', 'Total Profit (₹)']
segment_summary['Win Rate (%)'] = (segment_summary['Win Rate (%)'] * 100).round(2)

print("--- SEGMENT PERFORMANCE (INR) ---")
display(segment_summary)

# 3. Quick Plot to see the difference in Profit
segment_summary['Total Profit (₹)'].plot(kind='bar', color=['darkred', 'darkblue'], figsize=(8,5))
plt.title("Total Profit (₹) by Trade Size", fontsize=14)
plt.ylabel("Rupees (₹)")
plt.xticks(rotation=0)
plt.show()
```

```python
# Use 'trade_segment' and the correct PnL column name
segment_profit = merged.groupby('trade_segment')['Closed PnL INR'].mean()

print("--- Average Profit by Trade Size (INR) ---")
print(segment_profit.map('₹{:,.2f}'.format))

# Plot it so it's easy to see
segment_profit.plot(kind='bar', color=['#3498db', '#e74c3c'])
plt.title("Average Profit (INR) by Trade Segment")
plt.ylabel("Profit (₹)")
plt.show()
```

```python
# The "Ultimate Summary" Table
summary = merged.groupby(['classification', 'trade_segment']).agg({
    'Closed PnL INR': 'sum',
    'win': 'mean'
}).reset_index()

# Clean up the names for the report
summary.columns = ['Market Mood', 'Trade Size', 'Total Profit (₹)', 'Win Rate %']
summary['Win Rate %'] = (summary['Win Rate %'] * 100).round(1)
summary['Total Profit (₹)'] = summary['Total Profit (₹)'].round(2)

display(summary)
```

## Part C: Final Actionable Insights & Strategic Rules

Based on the statistical analysis of trade performance across different market sentiments, I propose the following rules for the Hyperliquid trading desk:

### 1. The "Euphoria" Risk Filter (Greed Strategy)
* **Finding:** Analysis shows that during "Greed" and "Extreme Greed," trade sizes increase by over 20%, but the Win Rate drops. This indicates FOMO-driven decision making.
* **Rule:** When the Fear & Greed Index is > 70, implement a **Hard Leverage Cap**. Reduce maximum position size by 30% to protect the account from the high-volatility "outliers" observed in the Greed distribution.

### 2. The "Fear" Selective Aggression (Fear Strategy)
* **Fiding:** Win rates are statistically higher during "Fear" days, suggesting that the trader identifies better entry points when the market is oversold.
* **Rule:** During "Fear" days, prioritize **high-conviction 'Small Trade' setups**. This allows for capital accumulation during market bottoms with a higher probability of success.

### 3. Professional Conclusion
By aligning position sizing with market sentiment rather than trading with a fixed size, a trader can significantly smooth their equity curve and reduce the "PnL swings" discovered in this data audit.
